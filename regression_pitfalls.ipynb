```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
N = 2000

C = np.random.randn(N,1)
A = 0.8 * np.random.randn(N,1)
K = A + 0.1 * np.random.randn(N,1)
X = C - 2*A + 0.2 * np.random.randn(N,1)
F = 3*X + 0.8 * np.random.randn(N,1)
D = -2*X + 0.5 * np.random.randn(N,1)
G = D + 0.5 * np.random.randn(N,1)
Y = 2*K - D + 0.2 * np.random.randn(N,1)
H = 0.5*Y + 0.1 * np.random.randn(N,1)
```

```python
predictors = X
outcome = D

predictors = sm.add_constant(predictors)
reg = sm.OLS(outcome, predictors).fit()
coefficients = reg.params
plt.plot(np.linspace(-10,10, len(predictors)),
         coefficients[1]*np.linspace(-10,10, len(predictors)),color='k')
plt.scatter(predictors[:,1],outcome)
plt.show()
print(coefficients)
```

```python
predictors = np.concatenate((D,K),1)
outcome = Y

predictors = sm.add_constant(predictors)
reg = sm.OLS(outcome, predictors).fit()
coefficients = reg.params
plt.plot(np.linspace(-10,10, len(predictors)),
         coefficients[1]*np.linspace(-10,10, len(predictors)),color='k')
plt.scatter(predictors[:,1],outcome)
plt.show()
print(coefficients)
```
